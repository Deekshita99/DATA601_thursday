1. Append to the dataframe: projected GDP (nominal) from IMF estimates from 2022 to 2027. (If data is not available for the country use None)
1. Add a new column and calculate average increase in terms of percentage. 
1. Sort by this column in descending order.

```python
#1.reading/loading the csv file created in lab
import pandas as pd
import requests
df = pd.read_csv("country.csv",index_col = 'Unnamed: 0')
```

```python
df
```

```python
#2.Scraping the data from given website
from bs4 import BeautifulSoup
url = 'https://en.wikipedia.org/wiki/List_of_countries_by_past_and_projected_GDP_(nominal)#IMF_projections_for_2020_through_2027'
page = requests.get(url)

soup = BeautifulSoup(page.content, 'html.parser')

table = soup.find_all('table')
df1 = pd.read_html(str(table))[5]
```

```python
df1
```

```python
df1.rename(columns={'Country (or dependent territory)': 'country_name'}, inplace=True)
```

```python
df1
```

```python
#3.merging two tables
countries_df = pd.merge(df1,df,on = "country_name", how = 'left')
countries_df.head()
```

```python
import numpy as np
countries_df = countries_df.replace(np.nan, 0)
countries_df
```

```python
#4.Adding new column and calculating average increase in terms of percentage.
countries_df['average'] = ((countries_df['2027']-countries_df['2022'])*100)/countries_df['2022']
countries_df
```

```python
#5.Sorting by average column in descending order.
countries_df.sort_values(by=['average'], ascending = False)
```
